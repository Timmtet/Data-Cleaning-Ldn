```python
import pandas as pd
from datetime import datetime, timedelta

# Assuming your dataset is stored in a pandas DataFrame called 'om'
# and the columns are named 'Current_date' and 'Listing_Date'

# Convert 'Current_date' column to datetime type
om['Current_date'] = pd.to_datetime(om['Current_date'], format='%d/%m/%Y')

# Extract the sign ('<' or '>') from 'Listing_Date' column
om['Sign'] = om['Listing_Date'].str.extract(r'([<>])')

# Extract the numeric value from 'Listing_Date' column
om['Days'] = om['Listing_Date'].str.extract(r'[<>]([0-9]+)').astype(float)

# Calculate the new 'Date_main' column based on different conditions
om['Date_main'] = om.apply(
    lambda row: (
        (row['Current_date'] - timedelta(days=row['Days'])).strftime('%Y-%m-%d')
        if pd.notnull(row['Days']) and pd.notnull(row['Sign']) and row['Sign'] in ['>', '<']
        else (row['Current_date'] - timedelta(days=row['Days']) + timedelta(days=1)).strftime('%Y-%m-%d')
        if row['Listing_Date'].endswith('yesterday') and pd.notnull(row['Days'])
        else row['Current_date'].strftime('%Y-%m-%d')
    ),
    axis=1
)

# Convert the 'Date_main' column back to string and add the sign ('<' or '>') from 'Listing_Date'
om['Date_main'] = om.apply(
    lambda row: (
        f"{row['Sign']} {row['Date_main']}"
        if pd.notnull(row['Sign']) and row['Sign'] in ['>', '<']
        else (row['Current_date'] - timedelta(days=1)).strftime('%Y-%m-%d')
        if row['Listing_Date'].endswith('yesterday')
        else row['Current_date'].strftime('%Y-%m-%d')
    ),
    axis=1
)

# Remove the 'Sign' and 'Days' columns if desired
om.drop(['Sign', 'Days'], axis=1, inplace=True)
```
