### i) Cleaning the Listing_Date Column

Here are the values available in the Listing_Date

```python
om.Listing_Date.value_counts()
```

To ensure uniformity with the dataset from other websites, we need to convert the values into a Date datatype. We will accomplish this by utilizing the 'Current_date' column, which represents the date the data was web scraped, and the 'Listing_date' column to compute a new column that will serve as our standardized listing date.

E.g If current date = 15/05/2023 and Listing date = OnTheMarket > 14 days. Date_main will be  01/05/2023

If current date = 15/05/2023 and Listing date = OnTheMarket today. Date_main will be 15/05/2023

```python
import pandas as pd
from datetime import datetime, timedelta

# Assuming your dataset is stored in a pandas DataFrame called 'om'
# and the columns are named 'Current_date' and 'Listing_Date'

# Convert 'Current_date' column to datetime type
om['Current_date'] = pd.to_datetime(om['Current_date'], format='%d/%m/%Y')

# Calculate the new 'Date_main' column based on different conditions
om['Date_main'] = om.apply(
    lambda row: (
        (row['Current_date']).strftime('%d/%m/%Y')
        if row['Listing_Date'] in ['Reduced today', 'OnTheMarket today']
        else (row['Current_date'] - timedelta(days=1)).strftime('%d/%m/%Y')
        if row['Listing_Date'] in ['Reduced yesterday', 'OnTheMarket yesterday']
        else (row['Current_date'] - timedelta(days=14)).strftime('%d/%m/%Y')  # > (current date - 14 days)
        if row['Listing_Date'] in ['OnTheMarket > 14 days', 'Reduced > 14 days']
        else (row['Current_date'] - timedelta(days=14)).strftime('%d/%m/%Y')  # < (current date - 14 days)
        if row['Listing_Date'] in ['OnTheMarket < 14 days', 'Reduced < 14 days']
        else (row['Current_date'] - timedelta(days=7)).strftime('%d/%m/%Y')  # < (current date - 7 days)
        if row['Listing_Date'] in ['OnTheMarket < 7 days', 'Reduced < 7 days']
        else None  # Add any other conditions or default value if needed
    ),
    axis=1
)

# Convert the column to datetime type
om['Date_main'] = pd.to_datetime(om['Date_main'])

# Convert the values to the desired format
om['Date_main'] = om['Date_main'].dt.strftime('%d/%m/%y')
```

### i) Cleaning the Listing_Date column

We will attempt to clean the Listing_Date to remove the 'Listed on' string attached to the listed date

```python
# Strip "Listed on" from Date values while preserving 
zp['Listing_Date'] = zp['Listing_Date'].str.replace('Listed on', '').str.strip()
```

```python
zp.head()
```

Now, we will convert the Listing dates into the format dd/mm/yyyy, for uniformity with the other datasets

```python
# Convert the date format to 'dd/mm/yyyy'
zp['Listing_Date'] = zp['Listing_Date'].str.replace('th', '').str.replace('st', '').str.replace('nd', '').str.replace('rd', '').str.replace(' ', '/')

# Convert the month name to the month number
try: 
    zp['Listing_Date'] = pd.to_datetime(zp['Listing_Date'], format='%d/%B/%Y').dt.strftime('%d/%m/%Y')
except ValueError:
    pass

# Convert the month name to the month number
try: 
    # Update the month name to month number conversion
    zp['Listing_Date'] = zp['Listing_Date'].str.replace('/Jan/', '/01/').str.replace('/Feb/', '/02/').str.replace('/Mar/', '/03/').str.replace('/Apr/', '/04/').str.replace('/May/', '/05/').str.replace('/Jun/', '/06/').str.replace('/Jul/', '/07/').str.replace('/Aug/', '/08/').str.replace('/Sep/', '/09/').str.replace('/Oct/', '/10/').str.replace('/Nov/', '/11/').str.replace('/Dec/', '/12/')
except ValueError:
    pass
```
